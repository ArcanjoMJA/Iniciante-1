```python
!pip install python-bcb
from bcb import sgs
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib


%matplotlib inline
matplotlib.style.use('bmh')
#url=#
#df = sgs.get({'IPCA': 433}, start='2002-02-01')
df = sgs.get({'Producao Industrial': 21859}, start='2002-02-01')
#df = sgs.get({'INCC': 7456}, start='2002-02-01')
#df = sgs.get({'IPCA Alimentos e Bebidas': 1635}, start='2002-02-01')
#df = sgs.get({'IPCA15': 7478}, start='2002-02-01')

#df = sgs.get({'IGP-M': 189}, start='2002-02-01')
#df = sgs.get({'RNDBF': 29037}, start='2002-02-01')

df.index = df.index.to_period('M')

```

```python
df.plot()
```

```python
import pmdarima as pm
import pandas as pd
#https://www.delftstack.com/howto/python/auto-arima-python/

#https://alkaline-ml.com/pmdarima/modules/generated/pmdarima.arima.auto_arima.html
```

```python
model_1 = pm.auto_arima(df,
                      seasonal=True,
                      trace=True,
                      error_action='ignore',
                      suppress_warnings=True,
                      stepwise=True)

print(model_1.summary())

print(model_1.summary())
```

```python
forecasts = model_1.predict(24)
```
